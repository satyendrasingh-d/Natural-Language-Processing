# Apply NER to identify entities such as names, organizations, and locations in a given text. Perform sentence segmentation on a paragraph and explain its importance in NLP tasks.

```python
import spacy
import pandas as pd
from collections import Counter
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text="data science has become an essential catalyst for global innovation by converting staggering amounts of daily-generated data—projected to exceed 463 exabytes daily—into strategic knowledge. The field operates through a structured workflow known as the data science life cycle, which begins by defining complex business problems and moves through critical stages like data scrubbing, exploratory analysis, and advanced machine learning modeling."
```

```python
print("Total Words:", len(text.split()))
```

```python
doc = nlp(text)
```

```python
entities = []
for ent in doc.ents:
    entities.append([ent.text, ent.label_])

# Convert to DataFrame
df_ner = pd.DataFrame(entities, columns=["Entity", "Label"])
entity_counts = Counter(df_ner["Label"])

print("\nEntity Type Counts:")
for label, count in entity_counts.items():
    print(f"{label}: {count}")
print("\nSample Named Entities:")
print(df_ner.head(25))
```
