# Que:1. Create a Doc object from the file owlcreek.txt

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
with open("owlcreek.txt","r") as file:
    text = file.read()
```

```python
doc = nlp(text)
```

# Que:2. How many tokens are contained in the file 

```python
token = len(doc)
print(f'Number of Tokens:{token}')
```

# Que:3. How many sentences are contained in the file

```python
sent = len(list(doc.sents))
print(sent)
```

# Que:4. Print the second sentence in the documnent

```python
sent = list(doc.sents)
```

```python
if len(sent) >= 2:
    print('Second sentence in the document:')
    print(sent[1].text)
```

# Que:5. For each token in the sentence above, print its text, POS tag, dep tag and lemma.

```python
for token in sent[1]:
    print(f'{token.text:{15}} {token.pos_:{10}} {token.dep_:{7}} {token.lemma_}')
```

# Que:6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text.

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

# Create patterns

```python
pattern = [{'LOWER':'swimming'},{'IS_SPACE': True, 'OP':'*'},{'LOWER': 'vigorously'}]
matcher.add('Swimming', [pattern])
```

```python
matches = matcher(doc)
matches
```
