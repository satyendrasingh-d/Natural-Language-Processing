```python
# Perform standared imports
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# From Spacy Basics:
doc = nlp(u'This is the first sentence. This is another sentence. This is the last sentence.')
for sent in doc.sents:
    print(sent)
```

# Adding Rules

```python
# Passing the segmentation start tokens happens during the nlp pipeline
doc2 = nlp(u'This is a sentence. This is a sentence. This is a sentence. ')
for token in doc2:
    print(token.is_sent_start,' '+token.text)
```

```python
# Spacy's default behavior 
doc3 = nlp(u'"Management is doint things right; leadership is doing the right things ." -Peter Drucker')
for sent in doc3.sents:
    print(sent)
```

```python
# Add a new role to the pipeline
from spacy.language import Language

@Language.component('set_custom_boundaries')
def set_custom_boundaries(doc):
    for token in doc[:-1]:
        if token.text ==';':
            doc[token.i+1].is_sent_start = True
    return doc
nlp.add_pipe("set_custom_boundaries", before='parser')
```

```python
# Re-run the doc object creation:
doc4 = nlp(u'"Management is doint things right; leadership is doing the right things ." -Peter Drucker')
for sent in doc4.sents:
    print(sent)
```
